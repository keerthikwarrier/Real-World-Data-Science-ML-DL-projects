### **2. Exploratory Data Analysis**
a. Perform an in-depth exploratory data analysis on the number of events by date, per year, and per day for each state and country

```python
Tesla['date']=pd.to_datetime(Tesla['date']) #Convert Date to DataFrame
#Extract day and year from Date Column
Tesla['year']=Tesla['date'].dt.year
Tesla['day']=Tesla['date'].dt.day
#Events per year
events_per_year = Tesla['year'].value_counts().sort_index()
print("\nEvents Per Year:",events_per_year)
#Events per day
events_per_day = Tesla['day'].value_counts().sort_index()
print("\nEvents Per Day:",events_per_day)
#Events per Day for each State
events_per_day_state = Tesla.groupby(['state', 'day']).size().unstack(fill_value=0)
print(events_per_day_state.head())
events_per_day_country = Tesla.groupby(['country', 'day']).size().unstack(fill_value=0)
print(events_per_day_country.head())
```

b.Analyze the different aspects of the death events. For example:

• What is the number of victims (deaths) in each accident?

• How many times did tesla drivers die?

• What is the proportion of events in which one or more occupants died?

• What is the distribution of events in which the vehicle hit a cyclist or a pedestrian?

• How many times did the accident involve the death of an occupant or driver of a Tesla along with a cyclist or pedestrian?

• What is the frequency of Tesla colliding with other vehicles?

```python
#number of victims (deaths) in each accident
number_victims_per_accident = Tesla['deaths'].value_counts().sort_index()

print("\n number of victims (deaths) in each accident:",number_victims_per_accident)
```

```python
# Count Tesla driver deaths
print(Tesla['tesladriver'].unique())

tesla_driver_deaths = Tesla['tesladriver'].value_counts()
print("\n tesla_driver_deaths:",tesla_driver_deaths)
```

```python
# Events with one or more Tesla occupants dead
tesla_occupant_death_events = Tesla['teslaoccupant'].str.lower().value_counts()
print("\n tesla_occupant_death_events:",tesla_occupant_death_events)
```

```python
# Events involving cyclists or pedestrians
cyclist_pedestrian_events = Tesla['cyclists/peds'].str.lower().value_counts()
cyclist_pedestrian_events
```

```python
# Events where Tesla driver or occupant died AND a cyclist/pedestrian was hit
joint_events = Tesla[
    (Tesla['cyclists/peds'].str.lower() == 'yes') &
    (
        (Tesla['tesladriver'].str.lower() == 'yes') |
        (Tesla['teslaoccupant'].str.lower() == 'yes')
    )
]
joint_events
```

```python
# Frequency of Tesla colliding with other vehicles
collisions_with_other_vehicles = Tesla['othervehicle'].str.lower().value_counts()
collisions_with_other_vehicles
```

c. Study the event distribution across models

```python
# Event distribution by Tesla model
model_counts = Tesla['model'].value_counts().sort_values(ascending=False)

# Plot the distribution
plt.figure(figsize=(10, 6))
model_counts.plot(kind='bar', color='steelblue')
plt.title('Event Distribution Across Tesla Models')
plt.xlabel('Tesla Model')
plt.ylabel('Number of Events')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

d.Check the distribution of verified Tesla autopilot deaths

```python
# Clean and inspect the column
Tesla['verifiedteslaautopilotdeaths'] = pd.to_numeric(
    Tesla['verifiedteslaautopilotdeaths'], errors='coerce'
)

# Drop NaNs for cleaner analysis
verified_ap_deaths = Tesla['verifiedteslaautopilotdeaths'].dropna()

# Plot distribution
plt.figure(figsize=(8, 5))
verified_ap_deaths.value_counts().sort_index().plot(kind='bar', color='crimson')
plt.title('Distribution of Verified Tesla Autopilot Deaths')
plt.xlabel('Number of Verified Autopilot Deaths in an Event')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
