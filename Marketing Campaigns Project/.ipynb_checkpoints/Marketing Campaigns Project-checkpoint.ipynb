#### 2. There are missing income values for some customers. Conduct missing value imputation, considering that customers with similar education and marital status tend to have comparable yearly incomes, on average. It may be necessary to cleanse the data before proceeding. Specifically, scrutinize the categories of education and marital status for data cleaning

```python
# Check unique values in "Education" and "Marital_Status"
education_unique = df["Education"].unique()
marital_status_unique = df["Marital_Status"].unique()
print(education_unique, marital_status_unique, df["Income"].isnull().sum())

#Standardize the category names of Education and Marital Status
# Standardize marital status categories
df["Marital_Status"] = df["Marital_Status"].replace({
    "YOLO": "Single",
    "Alone": "Single",
    "Absurd": "Single"  # Assuming these are unconventional ways of indicating being single
})

# Verify cleaned categories
print(df["Marital_Status"].unique())

# Compute mean income for each (Education, Marital_Status) group
income_means = df.groupby(["Education", "Marital_Status"])["Income"].mean()

# Impute missing income values based on the group mean
df["Income"] = df.apply(
    lambda row: income_means[row["Education"], row["Marital_Status"]] if pd.isnull(row["Income"]) else row["Income"],
    axis=1
)
print(income_means)
# Verify if all missing values are filled
df["Income"].isnull().sum()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Read the CSV and parse the date column
df = pd.read_csv("marketing_data.csv", parse_dates=True)

# Clean Income Column (if needed)
df['Income'] = df[' Income '].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Create Age Column
current_year = pd.to_datetime("today").year
df['Age'] = current_year - df['Year_Birth']

# Define Age Groups
def categorize_age(age):
    if age < 30:
        return '18-30'
    elif age < 50:
        return '31-50'
    else:
        return '51+'

df['Age_Group'] = df['Age'].apply(categorize_age)

# 1. ANOVA Test: Comparing Technological Proficiency (NumWebVisitsMonth) Across Age Groups
groups = [df[df['Age_Group'] == age]['NumWebVisitsMonth'] for age in df['Age_Group'].unique()]
anova_result = stats.f_oneway(*groups)
print("ANOVA Test for Web Visits across Age Groups:", anova_result)

# 2. Chi-Square Test: Shopping Preference (Online vs. In-Store) vs. Age Group
df['Shopping_Preference'] = np.where(df['NumWebPurchases'] > df['NumStorePurchases'], 'Online', 'In-Store')
contingency_table = pd.crosstab(df['Age_Group'], df['Shopping_Preference'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("\nChi-Square Test for Shopping Preference vs. Age Group:")
print(f"Chi2: {chi2}, P-value: {p}")

# Conclusion
alpha = 0.05
if anova_result.pvalue < alpha:
    print("\nTechnological proficiency (web visits) significantly differs across age groups.")
else:
    print("\nNo significant difference in technological proficiency across age groups.")

if p < alpha:
    print("Shopping preference is significantly related to age group.")
else:
    print("No significant relationship between shopping preference and age group.")
```

__b. Customers with children likely experience time constraints, making online shopping a more convenient option.__

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Read the CSV and parse the date column
df = pd.read_csv("marketing_data.csv", parse_dates=True)

# Clean Income Column (if needed)
df['Income'] = df[' Income '].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Create Age Column
df['Age'] = 2025 - df['Year_Birth']

# Define Age Groups
def categorize_age(age):
    if age < 30:
        return '18-30'
    elif age < 50:
        return '31-50'
    else:
        return '51+'

df['Age_Group'] = df['Age'].apply(categorize_age)

# Create 'Has_Children' column
df['Has_Children'] = np.where((df['Kidhome'] + df['Teenhome']) > 0, 'Yes', 'No')

# Categorize Shopping Preference
df['Shopping_Preference'] = np.where(df['NumWebPurchases'] > df['NumStorePurchases'], 'Online', 'In-Store')

# 1. ANOVA Test: Comparing Technological Proficiency (NumWebVisitsMonth) Across Age Groups
groups = [df[df['Age_Group'] == age]['NumWebVisitsMonth'] for age in df['Age_Group'].unique()]
anova_result = stats.f_oneway(*groups)
print("ANOVA Test for Web Visits across Age Groups:", anova_result)

# 2. Chi-Square Test: Shopping Preference (Online vs. In-Store) vs. Age Group
contingency_table = pd.crosstab(df['Age_Group'], df['Shopping_Preference'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("\nChi-Square Test for Shopping Preference vs. Age Group:")
print(f"Chi2: {chi2}, P-value: {p}")

# 3. Chi-Square Test: Shopping Preference vs. Having Children
contingency_children = pd.crosstab(df['Has_Children'], df['Shopping_Preference'])
chi2_child, p_child, dof_child, expected_child = stats.chi2_contingency(contingency_children)
print("\nChi-Square Test for Shopping Preference vs. Having Children:")
print(f"Chi2: {chi2_child}, P-value: {p_child}")

# 4. T-Test: Compare Online Purchases Between Parents and Non-Parents
ttest_result = stats.ttest_ind(df[df['Has_Children'] == 'Yes']['NumWebPurchases'],
                               df[df['Has_Children'] == 'No']['NumWebPurchases'],
                               equal_var=False)
print("\nT-Test for Online Purchases Between Customers with and without Children:", ttest_result)

# Conclusion
alpha = 0.05
if anova_result.pvalue < alpha:
    print("\nTechnological proficiency (web visits) significantly differs across age groups.")
else:
    print("\nNo significant difference in technological proficiency across age groups.")

if p < alpha:
    print("Shopping preference is significantly related to age group.")
else:
    print("No significant relationship between shopping preference and age group.")

if p_child < alpha:
    print("\nShopping preference is significantly related to having children.")
else:
    print("\nNo significant relationship between shopping preference and having children.")

if ttest_result.pvalue < alpha:
    print("Customers with children make significantly more online purchases.")
else:
    print("No significant difference in online purchases between customers with and without children.")
```

__c. Sales at physical stores may face the risk of cannibalization by alternative distribution channels.__
NOTE: Corporate cannibalism is when a product sees a decrease in sales volume or market share due to the release of some new product that has been introduced by the same company.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm

# Read the CSV and parse the date column
df = pd.read_csv("marketing_data.csv", parse_dates=True)

# Check for missing values in relevant columns
df = df[['NumWebPurchases', 'NumStorePurchases']].dropna()

# 1. Correlation Analysis
correlation = df[['NumWebPurchases', 'NumStorePurchases']].corr().iloc[0, 1]

# 2. Linear Regression (Predict Store Purchases based on Web Purchases)
X = df['NumWebPurchases']
Y = df['NumStorePurchases']
X = sm.add_constant(X)  # Add intercept
model = sm.OLS(Y, X).fit()

# 3. T-Test: Compare store purchases between high and low online shoppers
median_web_purchases = df['NumWebPurchases'].median()
high_online_shoppers = df[df['NumWebPurchases'] > median_web_purchases]['NumStorePurchases']
low_online_shoppers = df[df['NumWebPurchases'] <= median_web_purchases]['NumStorePurchases']
ttest_result = stats.ttest_ind(high_online_shoppers, low_online_shoppers, equal_var=False)

# Results
correlation, model.params[1], model.pvalues[1], ttest_result.pvalue
```

__d. Analyze the educational background of customers who lodged complaints in the last two years.__

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("marketing_data.csv")

# Convert 'Dt_Customer' to datetime (format based on the sample: month/day/year, e.g., '6/16/14')
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%m/%d/%y')

# Determine the latest customer date in the dataset (simulate current date)
max_date = df['Dt_Customer'].max()

# Define the start date for the "last two years" period
start_date = max_date - pd.DateOffset(years=2)

# Filter for customers who joined in the last two years
recent_customers = df[df['Dt_Customer'] >= start_date]

# Further filter for customers who lodged a complaint (assuming Complain==1 means they complained)
complaint_customers = recent_customers[recent_customers['Complain'] == 1]

# Count the educational background among these complaint customers
education_counts = complaint_customers['Education'].value_counts()

# Create a bar chart to visualize the distribution of education levels
plt.figure(figsize=(10, 6))
sns.barplot(x=education_counts.index, y=education_counts.values, palette='viridis')
plt.xlabel("Education Level")
plt.ylabel("Number of Customers")
plt.title("Educational Background of Complaint Customers in the Last Two Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
